```python
# Function to detect outliers using IQR
def detect_outliers(df1):
    Q1 = df1.quantile(0.25)
    Q3 = df1.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (df1 < lower_bound) | (df1 > upper_bound)

outliers = detect_outliers(df1.select_dtypes(include='number'))

# Show rows with outliers
outliers_rows = df1[outliers.any(axis=1)]
print(outliers_rows)
```
