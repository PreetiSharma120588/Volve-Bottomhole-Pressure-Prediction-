```python
# Ensure the key columns are numeric (convert if necessary)
key_columns = ['AVG_DOWNHOLE_PRESSURE','AVG_WHP_P','BORE_OIL_VOL']
df1[key_columns]=df1[key_columns].apply(pd.to_numeric,errors='coerce')

for column in key_columns:
    fig, axes=plt.subplots(1,3,figsize=(10,6))
    
# Histogram
    axes[0].hist(df1[column].dropna(),bins=30, color='skyblue',edgecolor='black')
    axes[0].set_title(f'Histogram of {column}')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('Frequency')
```
