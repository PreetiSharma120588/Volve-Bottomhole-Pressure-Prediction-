```python
# Grouping data by well name and calculating average pressure
grouped_data=df_daily.groupby('NPD_WELL_BORE_NAME')['AVG_DOWNHOLE_PRESSURE'].mean()

# Plotting a bar chart
plt.bar(grouped_data.index, grouped_data.values,color='skyblue',edgecolor='black')
plt.xlabel('Well Name')
plt.ylabel('Average Downhole Pressure')
plt.title('Average Downhole Pressure by Well')

# Adding labels on top of the bars
for i, value in enumerate(grouped_data.values):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45) 
plt.show()
```

```python
# Grouping data by well name and calculating average pressure
grouped_data=df_daily.groupby('NPD_WELL_BORE_NAME')['AVG_WHP_P'].mean()

# Plotting a bar chart
plt.bar(grouped_data.index, grouped_data.values,color='skyblue',edgecolor='black')
plt.xlabel('Well Name')
plt.ylabel('Average Wellhead Pressure')
plt.title('Average Wellhead Pressure by Well')

# Adding labels on top of the bars
for i, value in enumerate(grouped_data.values):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45) 
plt.show()
```

```python
# Grouping data by well name and calculating average pressure
grouped_data=df_daily.groupby('NPD_WELL_BORE_NAME')['AVG_WHT_P'].mean()

# Plotting a bar chart
plt.bar(grouped_data.index, grouped_data.values,color='skyblue',edgecolor='black')
plt.xlabel('Well Name')
plt.ylabel('Average Wellhead Temprature')
plt.title('Average Wellhead Temprature by Well')

# Adding labels on top of the bars
for i, value in enumerate(grouped_data.values):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45) 
plt.show()
```

```python
# Grouping data by well name and calculating average pressure
grouped_data=df_daily.groupby('NPD_WELL_BORE_NAME')['BORE_OIL_VOL'].mean()

# Plotting a bar chart
plt.bar(grouped_data.index, grouped_data.values,color='skyblue',edgecolor='black')
plt.xlabel('Well Name')
plt.ylabel('Average Oil Volume')
plt.title('Average Oil by each Well')

# Adding labels on top of the bars
for i, value in enumerate(grouped_data.values):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45) 
plt.show()
```
